```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv",encoding="utf-8")
```

```python
df.head(5)
```

```python
df.corr(numeric_only=True)
```

```python
df.drop(["User_ID"],axis=1,inplace=True)
```

```python
df.drop(["Product_ID"],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
model=LabelEncoder()
df["Gender"]=model.fit_transform(df["Gender"])
df["City_Category"]=model.fit_transform(df['City_Category'])
df["Stay_In_Current_City_Years"]=model.fit_transform(df['Stay_In_Current_City_Years'])
df['Marital_Status']=model.fit_transform(df['Marital_Status'])
```

```python
df.head(5)
```

```python
df['Product_Category_2'].fillna(df['Product_Category_2'].mean(),inplace=True)
```

```python
df["Product_Category_3"].fillna(df["Product_Category_3"].mean(),inplace=True)
```

```python
df.head(5)
```

```python
def convert_age_to_number(age_str):
    if "-" in age_str:
        start, end = map(int, age_str.split("-"))
        return (start + end/2) 
    elif "+" in age_str:
        return int(age_str[:-1])  
    else:
        return int(age_str) 

df["Age"] = df["Age"].apply(convert_age_to_number)
```

```python
df.head(10)
```

```python
df.corr()
```

```python
y=df["Purchase"]
x=df.drop(["Purchase"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1234)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
prediction=model.predict(x_test)
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,prediction)
```

```python
metrics.r2_score(y_test,prediction)
```
