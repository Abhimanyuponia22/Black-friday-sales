```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
model=LabelEncoder()
df["Gender"]=model.fit_transform(df["Gender"])
df["Age"]=model.fit_transform(df["Age"])
df["City_Category"]=model.fit_transform(df["City_Category"])
df["Product_Category_1"]=model.fit_transform(df["Product_Category_1"])
df["Product_Category_2"]=model.fit_transform(df["Product_Category_2"])
df["Product_Category_3"]=model.fit_transform(df["Product_Category_3"])
df["Stay_In_Current_City_Years"]=model.fit_transform(df["Stay_In_Current_City_Years"])
df['Occupation']=model.fit_transform(df['Occupation'])
```

```python
df.head(5)
```

```python
df.drop(["User_ID","Product_ID"],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df.corr()
```

```python
x=df.drop(["Purchase"],axis=1)
y=df["Purchase"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=12345)
```

```python
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(x_train,y_train)
```

```python
from sklearn import metrics
```

```python
pred_lr=model_lr.predict(x_test)
```

```python
pred_lr
```

```python
y_test
```

```python
metrics.mean_absolute_error(pred_lr,y_test)
```

```python
metrics.mean_squared_error(pred_lr,y_test)
```

```python
metrics.r2_score(y_test,pred_lr)
```

```python
import xgboost as xg
```

```python
model_xg=xg.XGBRegressor()
```

```python
model_xg.fit(x_train,y_train)
```

```python
pred_xg=model_xg.predict(x_test)
```

```python
metrics.r2_score(y_test,pred_xg)
```

```python
import catboost as cb
```

```python
model_cb=cb.CatBoostRegressor()
model_cb.fit(x_train,y_train)
```

```python
pred_cb=model_cb.predict(x_test)
```

```python
metrics.r2_score(y_test,pred_cb)
```
